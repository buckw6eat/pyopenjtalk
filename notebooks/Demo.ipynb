## Text-to-speech (TTS)

```python
x, sr = pyopenjtalk.tts("プリプリに脂の乗った寒ブリのお寿司が食べたい")
plt.figure(figsize=(16,4))
librosa.display.waveplot(x, sr, x_axis="time")
Audio(x, rate=sr)
```
